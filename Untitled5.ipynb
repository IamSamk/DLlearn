```python
import tensorflow as tf
import os, zipfile
```

```python
with zipfile.ZipFile("shells_pebbles.zip",'r') as r:
  r.extractall(".")
```

```python
base_dir = os.path.join("/content",'shells')
train_dir = "/content/train"
test_dir = "/content/test"
```

```python


from tensorflow.keras.preprocessing.image import ImageDataGenerator
size = (150,150)


train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

```

```python

train_gen = train_datagen.flow_from_directory(
    train_dir,
    target_size = size,
    batch_size = 20,
    class_mode = 'binary'
)
test_gen = test_datagen.flow_from_directory(
    test_dir,
    target_size = size,
    batch_size = 20,
    class_mode = 'binary'
)
```

```python
print(train_gen)
print(test_gen)
```

```python
from tensorflow.keras import models, layers


model = models.Sequential([
    layers.Conv2D(32,(3,3),activation='relu',input_shape=(150,150,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64,(3,3),activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128,(3,3),activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(256, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])


```

```python
model.compile(optimizer='adam',loss='binary_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(
    train_gen,
    validation_data = test_gen,
    epochs = 10
)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# 1. Grab a single image from the test generator
img_batch, label_batch = next(test_gen)
img = img_batch[0]
img_input = np.expand_dims(img, axis=0)

# 2. Saliency Map Logic
def compute_saliency(model, image):
    image_tensor = tf.convert_to_tensor(image)
    with tf.GradientTape() as tape:
        tape.watch(image_tensor)
        prediction = model(image_tensor)

    # Get gradients of the prediction with respect to the input image
    grads = tape.gradient(prediction, image_tensor)
    # Take the maximum absolute value across the RGB channels
    saliency = tf.reduce_max(tf.abs(grads), axis=-1)[0]
    # Normalize for visualization
    saliency = (saliency - tf.reduce_min(saliency)) / (tf.reduce_max(saliency) - tf.reduce_min(saliency))
    return saliency

# 3. Occlusion Sensitivity Logic
def occlusion_sensitivity(model, image, patch_size=15):
    h, w, _ = image.shape
    occ_map = np.zeros((h, w))
    # Get baseline prediction
    baseline_prob = model.predict(np.expand_dims(image, 0), verbose=0)[0][0]

    for y in range(0, h, patch_size):
        for x in range(0, w, patch_size):
            img_copy = image.copy()
            # "Occlude" a patch by making it black
            img_copy[y:y+patch_size, x:x+patch_size, :] = 0
            new_prob = model.predict(np.expand_dims(img_copy, 0), verbose=0)[0][0]
            # Measure how much the probability changed
            occ_map[y:y+patch_size, x:x+patch_size] = np.abs(baseline_prob - new_prob)
    return occ_map

# Execute and Plot
saliency = compute_saliency(model, img_input)
occ = occlusion_sensitivity(model, img)

fig, ax = plt.subplots(1, 3, figsize=(18, 6))
ax[0].imshow(img)
ax[0].set_title(f"Original (Label: {int(label_batch[0])})")
ax[1].imshow(saliency, cmap='jet')
ax[1].set_title("Saliency Map (Gradients)")
ax[2].imshow(occ, cmap='hot')
ax[2].set_title("Occlusion Heatmap")
plt.show()
```

```python
from lime import lime_image
from skimage.segmentation import mark_boundaries

explainer = lime_image.LimeImageExplainer()

# We define a prediction function that LIME can call
def predict_func(images):
    return model.predict(images, verbose=0)

# LIME expects a 3-channel output (even for binary), so we'll wrap our model
def model_predict_lime(img_list):
    preds = model.predict(img_list, verbose=0)
    # Convert sigmoid output [p] to [1-p, p] for LIME's multiclass expectation
    return np.hstack([1-preds, preds])

explanation = explainer.explain_instance(img.astype('double'),
                                         model_predict_lime,
                                         top_labels=1, hide_color=0, num_samples=1000)

temp, mask = explanation.get_image_and_mask(explanation.top_labels[0],
                                            positive_only=True, num_features=5, hide_rest=False)

plt.figure(figsize=(8,8))
plt.imshow(mark_boundaries(temp, mask))
plt.title("LIME: Areas Supporting the Prediction")
plt.axis('off')
plt.show()
```
